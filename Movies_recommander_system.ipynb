```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
# that use for converting dictinory to list 
```

```python
# for making to all row make function for thats 
def Convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
df['genres']=df['genres'].apply(Convert)
```

```python
df['keywords']=df['keywords'].apply(Convert)
```

```python
# for cast purpose we collecting 5 stars of all of them
def Convert2(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=5:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
df['cast']=df['cast'].apply(Convert2)
```

```python
# for crew purpose we collecting director of film
def Convert3(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
df['crew']=df['crew'].apply(Convert3)
```

```python
df
```

```python
df['overview']=df['overview'].apply(lambda x:x.split()) #converting into list
```

```python
df.head()
```

```python
# here we are complting to converting all dic to list 
# now we want to replace the space b/w word with nothing
df['genres']=df['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
df['keywords']=df['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
df['cast']=df['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
df
```

```python
df["Tages"]=df['genres']+df['cast']+df['crew']+df['overview']+df['keywords']
```

```python
df.iloc[0].Tages
```

```python
new_df=df[['movie_id','title','Tages']]
```

```python
new_df
```

```python
new_df["Tages"]=new_df["Tages"].apply(lambda x: " ".join(x))   # converting into list to string
```

```python
new_df['Tages']=new_df['Tages'].apply(lambda x:x.lower())   # converting into lower characters
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df["Tages"]=new_df["Tages"].apply(stem)
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vactors=cv.fit_transform(new_df['Tages']).toarray()
```

```python
vactors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity    # they measure angle between two vectors
similarity=cosine_similarity(vactors)    ## that is always between 1 and 0. 1 means too far and 0 means so close
```

```python
cosine_similarity(vactors).shape
```

```python
similarity
```

```python
new_df[new_df["title"]=='Avatar'].index[0]  ## that give you index of movie
```

```python
def recommand(movie):
    movie_index=new_df[new_df["title"]==movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:11]  # enumetrate function help for holding a index 
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommand("Avatar")
```

```python
recommand("Batman")
```
